```python
# requests를 통한 http 요청
import requests
```

```python
res = requests.get(url)
res
```

```python
# BeautifulSoup 사용해서 table 태그 찾기
from bs4 import BeautifulSoup as bs
```

```python
html = bs(res.text, "lxml")
```

```python
temp = html.select("table")
temp
```

```python
# pandas의 read_html 이용하여 dataframe 만들기
table = pd.read_html(str(temp))
table[0]
```

```python
df_temp = table[0]
df_temp
```

```python
# 페이지별 데이터 수집 함수
def get_info_list(page_no):
    url = f"https://finance.naver.com/item/news_notice.nhn?code=035720&page={page_no}"
    res = requests.get(url)
    html = bs(res.text, "lxml")
    temp = html.select("table")
    df_temp = pd.read_html(str(temp))[0]
    return df_temp
```

```python
# 함수 확인
page_no = 3
get_info_list(page_no)
```

```python
# 반복문 이용해 가져오고 싶은 페이지 데이터 수집
import time

page_no = 1
start_page = 1
end_page = 5

information_list = []
for page_no in range(start_page, end_page + 1):
    df_temp_info = get_info_list(page_no)
    information_list.append(df_temp_info)
    time.sleep(0.1)
    
```

```python
len(information_list)
```

```python
# 수집한 데이터 합치기
df = pd.concat(information_list)
```

```python
df.head()
```

```python
df.tail()
```

```python
# 최근 날짜 값으로 파일명 만들기
date = df.iloc[0]["날짜"]
date = date.replace(".", "-")
date
```

```python
file_name = f"kakao-info-{date}.csv"
```

```python
# 파일 저장
df.to_csv(file_name, index=False)
```
