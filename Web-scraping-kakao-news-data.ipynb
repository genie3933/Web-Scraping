# 카카오 뉴스 정보 데이터

```python
import pandas as pd
```

```python
# 수집할 url
url = "https://finance.naver.com/item/news_news.nhn?code=035720&page=1&sm=title_entity_id.basic&clusterId="
url
```

```python
import requests
# requests를 통한 html 가져오기
res = requests.get(url)
res
```

```python
res.text
```

```python
# bs로 받아온 html 문서 해석하기
from bs4 import BeautifulSoup as bs
html = bs(res.text, "lxml")
```

```python
# .select 사용해서 table 태그 찾기
temp = html.select("table")
temp
```

```python
# dataframe 형태로 만들기
table = pd.read_html(str(temp))
df_temp = table[0]
df_temp
```

```python
# 페이지 별 데이터 수집 함수 만들기
def get_kakao_news_list(page_no):
    url = f"https://finance.naver.com/item/news_news.nhn?code=035720&page={page_no}&sm=title_entity_id.basic&clusterId="
    res = requests.get(url)
    html = bs(res.text, 'lxml')
    temp = html.select("table")
    df_temp = pd.read_html(str(temp))[0]
    return df_temp
```

```python
# 함수 확인
page_no = 3
get_kakao_news_list(page_no)
```

```python
# 반복문을 이용해서 1~10 페이지 뉴스 리스트 만들기
import time

page_no = 1
start_page = 1
end_page = 10

news_list = []
for page_no in range(start_page, end_page + 1):
    news_list.append(get_kakao_news_list(page_no))
    time.sleep(0.1)
```

```python
# 가져온 데이터 합치기
df = pd.concat(news_list)
len(df)
```

```python
# 데이터 확인
df.head()
```

```python
df.tail()
```

```python
# 중복데이터 제거
print(df.shape)
df = df.drop_duplicates()
print(df.shape)
```

```python
# 최근 날짜 값으로 파일명 만들기
date = df.iloc[0]["날짜"].split(" ")[0]
date = date.replace(".", "-")
date
```

```python
file_name = f"kakao-news-{date}.csv"
```

```python
# 파일 저장
df.to_csv(file_name, index=False)
```
