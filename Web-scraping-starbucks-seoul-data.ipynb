# 서울 스타벅스 매장 위치정보 데이터 수집

```python
# 라이브러리 불러오기
import requests
import pandas as pd
```

```python
# 수집할 url, form_data 링크 가져오기
url = "https://www.starbucks.co.kr/store/getStore.do?r=PHQXAAQE4U"
params = "in_biz_cds=0&in_scodes=0&ins_lat=37.56682&ins_lng=126.97865&search_text=&p_sido_cd=01&p_gugun_cd=&isError=true&in_distance=0&in_biz_cd=&iend=1000&searchType=C&set_date=&rndCod=SCR6WY2P6S&all_store=0&T03=0&T01=0&T27=0&T12=0&T09=0&T30=0&T05=0&T22=0&T21=0&T10=0&T36=0&T43=0&T48=0&P10=0&P50=0&P20=0&P60=0&P30=0&P70=0&P40=0&P80=0&whcroad_yn=0&P90=0&new_bool=0"
```

```python
# requests로 html 문서 가져오기
res = requests.post(f"{url}?{params}")
res
```

```python
# json 형태의 타입 데이터를 새 변수에 저장
starbucks_seoul_store_info = res.json()['list']
```

```python
# 필요한 컬럼만 모아 데이터프레임 형태로 변경
df = pd.DataFrame(starbucks_seoul_store_info)[["s_code", "s_name", "tel", "sido_code", 
                                      "sido_name", "gugun_code", "gugun_name",
                                     "addr", "lat", "lot"]]
```
