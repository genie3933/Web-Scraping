```python
# 라이브러리 로드
import selenium
from selenium import webdriver
import time
import pandas as pd
from bs4 import BeautifulSoup
import requests
from itertools import repeat
import re
```

```python
# 크롬 드라이버 열기
driver = webdriver.Chrome('chromedriver.exe')
driver.implicitly_wait(5) # 대기 걸기
driver.maximize_window()
```

```python
# MAP OF THE SOUL : 7 앨범
# 각 컬럼별 데이터 리스트 선정 (7개)

release_date = [] # 발매일
artist = [] # 가수명
album_name = [] # 앨범명
song_name = [] # 곡 명
genre = [] # 장르
like_count = [] # 곡 좋아요 수
lyrics = [] # 가사


# 곡 명, 가수명, 앨범명, 발매일, 장르, 곡 좋아요 수, 가사 수집

for i in range(20): 
    # 드라이버가 해당 url 접속
    url = 'https://www.melon.com/album/detail.htm?albumId=10377346'
    driver.get(url)
    time.sleep(2)
    
    # 곡 정보 클릭
    driver.find_element_by_xpath(f'/html/body/div[1]/div[3]/div/div/div[3]/form/div/table/tbody/tr[{i + 1}]/td[3]/div/a').click()
    time.sleep(2)
    
    # 발매일 수집
    release_date.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div/div/form/div/div/div[2]/div[2]/dl/dd[2]').text)
    # 가수명 수집
    artist.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div/div/form/div/div/div[2]/div[1]/div[2]/a/span[1]').text)
    # 앨범명 수집
    album_name.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div/div/form/div/div/div[2]/div[2]/dl/dd[1]/a').text)
    # 곡 명 수집
    song_name.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div/div/form/div/div/div[2]/div[1]/div[1]').text)
    # 장르 수집
    genre.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div/div/form/div/div/div[2]/div[2]/dl/dd[3]').text)
    # 곡 좋아요 수 수집
    like_count.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div/div/form/div/div/div[2]/div[3]/button[1]/span[2]').text)
    # 가사 수집
    lyrics.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div/div/div[2]/div[2]/div').text)
    
```
